```python
import cv2
import numpy as np

# Cargar la imagen multiespectral
ruta_imagen = "E:\\prueba2\\paso1\\UCLV\\NIR\\IMG_180129_174913_0116_NIR.TIF"

imagen = cv2.imread(ruta_imagen)

# Convertir la imagen a escala de grises
imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

# Aplicar la umbralización para obtener la máscara
umbral = 100
_, mascara = cv2.threshold(imagen_gris, umbral, 255, cv2.THRESH_BINARY)

# Mostrar la máscara
cv2.imshow("Máscara", mascara)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Guardar la máscara en un archivo
ruta_mascara = "E:\\prueba2\\paso1\\UCLV\\mascara\\mascara.tif"
cv2.imwrite(ruta_mascara, mascara)
```

```python
import cv2
import numpy as np

# Cargar la imagen multiespectral
ruta_imagen = "E:\\prueba2\\paso1\\UCLV\\NIR\\IMG_180129_174913_0116_NIR.TIF"
imagen = cv2.imread(ruta_imagen)

# Convertir la imagen a escala de grises
imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

# Aplicar la segmentación por umbralización adaptativa para obtener la máscara
mascara = cv2.adaptiveThreshold(imagen_gris, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

# Mostrar la máscara
cv2.imshow("Máscara", mascara)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Guardar la máscara en un archivo
ruta_mascara ="E:\\prueba2\\paso1\\UCLV\\mascara\\mascara1.tif"
cv2.imwrite(ruta_mascara, mascara)
```

```python
import numpy as np
from skimage import io, filters

# Cargar la imagen multiespectral
ruta_imagen = "E:\\prueba2\\paso1\\UCLV\\NIR\\IMG_180129_174913_0116_NIR.TIF"
imagen = io.imread(ruta_imagen)

# Convertir la imagen a escala de grises
imagen_gris = imagen.astype(np.uint8)

# Aplicar un umbral utilizando el método de Otsu
umbral = filters.threshold_otsu(imagen_gris)
mascara = imagen_gris > umbral

# Mostrar la máscara
io.imshow(mascara)
io.show()

# Guardar la máscara en un archivo
ruta_mascara = "E:\\prueba2\\paso1\\UCLV\\mascara\\mascara2.tif"
io.imsave(ruta_mascara, mascara.astype(np.uint8))
```

```python
import numpy as np
from skimage import io

# Cargar la imagen multiespectral
ruta_imagen = "E:\\prueba2\\paso1\\UCLV\\NuevoNIR.tif"
imagen = io.imread(ruta_imagen)

# Convertir la imagen a escala de grises
imagen_gris = np.mean(imagen, axis=2).astype(np.uint16)

# Definir los valores de umbral
umbral_min = 1000
umbral_max = 2000

# Aplicar la umbralización
mascara = np.logical_and(imagen_gris > umbral_min, imagen_gris < umbral_max)

# Mostrar la máscara
io.imshow(mascara)
io.show()

# Guardar la máscara en un archivo
ruta_mascara = "E:\\prueba2\\paso1\\UCLV\\mascara\\mascara4.tif"
io.imsave(ruta_mascara, mascara.astype(np.uint8))
```
