```python
import tifffile
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import cv2
```

```python
# Ruta de la imagen TIFF 
ruta_imagen1 = "dnir.tif"

# Carga la imagen TIFF
nir_image = tifffile.imread(ruta_imagen1)

# Obtener las dimensiones de la imagen
height, width = nir_image.shape

# Reshape la imagen para tener una matriz de píxeles
pixels = nir_image.reshape(-1, 1)

# Obtener la posición (fila, columna) de cada píxel junto con su valor
pixel_positions = []
for (fila, columna), valor in np.ndenumerate(pixels):
    if valor != 65535:
        pixel_positions.append(fila)
        
filter_pixels = pixels != 65535

clean_pixel = pixels[filter_pixels].reshape(-1, 1)
```

# Kmeans

```python
# Definir el número de clusters (2 en este caso para separar cultivo y suelo)
n_clusters = 2

# Aplicar K-Means para agrupar los píxeles
kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=0)
kmeans.fit(clean_pixel)

# Obtener las etiquetas de los clusters asignados a cada píxel
labels = kmeans.labels_

# Asignar etiquetas binarias (0 para cultivo y 65535 para suelo)
segmented_image = np.zeros_like(clean_pixel, dtype=np.uint16)
cultivo_label = np.argmin(np.bincount(labels))  # Determinar la etiqueta del cultivo
segmented_image[labels == cultivo_label] = 65535  # Asignar 65535 a los píxeles del cultivo
```

```python
# Crear una nueva matriz llena de 65535 para la segmentación
new_nir_image = np.full((height, width), 65535, dtype=np.uint16)

# Reshape la imagen para tener una matriz de píxeles
pixels1 = new_nir_image.reshape(-1, 1)

# Asignar los valores limpios a las posiciones correspondientes
for i in range(len(pixel_positions)):
    fila = pixel_positions[i]
    pixels1[fila] = segmented_image[i]
    
segmented_image1 = pixels1.reshape(height, width)
```

```python
wss = 0.0
etiquetas = np.where(segmented_image == 65535, 1, 0)
# Calcular los nuevos centroides como el promedio de los puntos asignados a cada cluster
for k in range(2):
    centroides[k] = np.mean(clean_pixel[etiquetas == k])
    wss += np.sum(np.abs(clean_pixel[etiquetas == k] - centroides[k]) ** 2)
wss
```

```python
# Mostrar la imagen con matplotlib
plt.imshow(segmented_image1, cmap='gray')
plt.title('Imagen con Valores Limpios')
plt.axis('off')  # Oculta los ejes
plt.show()
```
