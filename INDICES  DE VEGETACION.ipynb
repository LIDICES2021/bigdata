```python
import numpy as np
import rasterio

# Rutas de las imágenes en cada banda
green_band_path = 'GRE_banda3.tif'
nir_band_path = 'NIR_banda1.tif'

# Abrir las imágenes multiespectrales
with rasterio.open(green_band_path) as green_band:
    green = green_band.read(1).astype('float32')  # Leer la banda verde

with rasterio.open(nir_band_path) as nir_band:
    nir = nir_band.read(1).astype('float32')  # Leer la banda infrarroja cercana

# Calcular el NDWI
ndwi = (green - nir) / (green + nir)

# Normalizar el rango del EVI a valores entre 0 y 255
ndwi_normalized = ((ndwi - np.min(ndwi)) / (np.max(ndwi) - np.min(ndwi))) * 255


# Guardar el resultado en una imagen
cv2.imwrite("ndwi_imageM.tif", ndwi_normalized.astype(np.uint8))

print("Índice NDWI calculado y guardado exitosamente.")


# Mostrar el EVI
plt.imshow(ndwi_normalized, cmap='RdYlGn')
plt.colorbar()
plt.show()
```

```python
import numpy as np
import rasterio

# Rutas de las imágenes en cada banda
red_band_path = 'RED.tif'
nir_band_path = 'NIR.tif'

# Abrir las imágenes multiespectrales
with rasterio.open(red_band_path) as red_band:
    red = red_band.read(1).astype('float32')  # Leer la banda roja

with rasterio.open(nir_band_path) as nir_band:
    nir = nir_band.read(1).astype('float32')  # Leer la banda infrarroja cercana

# Calcular el MSAVI
msavi = (2 * nir + 1 - np.sqrt((2 * nir + 1) ** 2 - 8 * (nir - red))) / 2

# Guardar el resultado en un archivo de imagen
output_path = 'msavi.tif'
with rasterio.open(output_path, 'w', driver='GTiff', width=red_band.width, height=red_band.height, count=1, dtype=msavi.dtype) as output:
    output.write(msavi, 1)  # Escribir el resultado en el archivo

print("Índice MSAVI calculado y guardado exitosamente.")

# Mostrar el EVI
plt.imshow(msavi, cmap='RdYlGn')
plt.colorbar()
plt.show()

```

```python
import numpy as np
import rasterio

# Rutas de las imágenes en cada banda
red_band_path = 'RED.tif'
nir_band_path = 'NIR.tif'

# Abrir las imágenes multiespectrales
with rasterio.open(red_band_path) as red_band:
    red = red_band.read(1).astype('float32')  # Leer la banda roja

with rasterio.open(nir_band_path) as nir_band:
    nir = nir_band.read(1).astype('float32')  # Leer la banda infrarroja cercana

# Calcular el OSAVI
osavi = (nir - red) / (nir + red + 0.16)


# Normalizar el rango del EVI a valores entre 0 y 255
osavi_normalized = ((msavi - np.min(msavi)) / (np.max(msavi) - np.min(msavi))) * 255


# Guardar el resultado en una imagen
cv2.imwrite("osavi_imageM.tif", osavi_normalized.astype(np.uint8))

# Mostrar el EVI
plt.imshow(osavi_normalized, cmap='RdYlGn')
plt.colorbar()
plt.show()

print("Índice OSAVI calculado y guardado exitosamente.")
```
