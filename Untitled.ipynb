```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el índice de vegetación (por ejemplo, NDVI)
ndvi = np.load('ndvi_result.npy')

# Crear un mapa de colores personalizado para resaltar el índice de vegetación
cmap = sns.diverging_palette(150, 275, s=80, l=80, n=256, as_cmap=True)

# Mostrar el índice de vegetación utilizando el mapa de colores personalizado
sns.heatmap(ndvi, cmap=cmap, center=0, cbar=True, cbar_kws={'label': 'NDVI'})

# Personalizar los ejes y el título si es necesario
plt.xlabel('Columnas')
plt.ylabel('Filas')
plt.title('NDVI Result')

# Mostrar la figura
plt.show()
```
