```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Ruta del directorio que contiene las imágenes
directorio = r'C:\Users\lidices\Desktop\proyecto\codigos\imagen'

# Obtener la lista de nombres de archivo de las imágenes en el directorio
nombres_archivos = os.listdir(directorio)

# Array para almacenar los valores de frecuencia del histograma
frecuencias = np.zeros(2**16, dtype=int)

# Recorrer cada imagen y calcular el histograma del canal NIR
for nombre_archivo in nombres_archivos:
    if nombre_archivo.endswith('.TIF'):
        ruta_imagen = os.path.join(directorio, nombre_archivo)
        imagen = Image.open(ruta_imagen)
        canal_nir = np.array(imagen, dtype=np.uint16)  # Leer la imagen completa como el canal NIR
        hist, _ = np.histogram(canal_nir, bins=2**16, range=(0, 2**16-1))
        frecuencias += hist

# Visualizar el histograma
niveles_intensidad = np.arange(2**16)
plt.plot(niveles_intensidad, frecuencias)
plt.xlabel('Niveles de Intensidad')
plt.ylabel('Frecuencias')
plt.title('Histograma del Canal NIR')
plt.show()
```
