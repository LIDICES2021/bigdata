```python
import numpy as np
import pandas as pd
import rasterio

# Cargar las imágenes de las bandas espectrales
nir_band_path = "dnir.tif"
red_band_path = "dred.tif"
green_band_path = "dgre.tif"
red_edge_band_path = "dreg.tif"

# Cargar los índices de vegetación
ndvi_path = "ndvi_result.tif"
savi_path = "savi_result.tif"
evi2_path = "evi2_result.tif"
ndre_path = "ndre_result.tif"
gndvi_path = "gndvi_result.tif"

# Leer los datos de las bandas espectrales
with rasterio.open(nir_band_path) as nir_band_img:
    nir_band = nir_band_img.read(1)

with rasterio.open(red_band_path) as red_band_img:
    red_band = red_band_img.read(1)

with rasterio.open(green_band_path) as green_band_img:
    green_band = green_band_img.read(1)

with rasterio.open(red_edge_band_path) as red_edge_band_img:
    red_edge_band = red_edge_band_img.read(1)

# Leer los datos de los índices de vegetación
with rasterio.open(ndvi_path) as ndvi_img:
    ndvi = ndvi_img.read(1)

with rasterio.open(savi_path) as savi_img:
    savi = savi_img.read(1)

with rasterio.open(evi2_path) as evi2_img:
    evi2 = evi2_img.read(1)

with rasterio.open(ndre_path) as ndre_img:
    ndre = ndre_img.read(1)

with rasterio.open(gndvi_path) as gndvi_img:
    gndvi = gndvi_img.read(1)

# Eliminar los píxeles con valor 65535
nir_band = np.where(nir_band == 65535, 0, nir_band)
red_band = np.where(red_band == 65535, 0, red_band)
green_band = np.where(green_band == 65535, 0, green_band)
red_edge_band = np.where(red_edge_band == 65535, 0, red_edge_band)
ndvi = np.where(ndvi == 65535, 0, ndvi)
savi = np.where(savi == 65535, 0, savi)
evi2 = np.where(evi2 == 65535, 0, evi2)
ndre = np.where(ndre == 65535, 0, ndre)
gndvi = np.where(gndvi == 65535, 0, gndvi)

# Crear un DataFrame con las características
data = {
    'nir_band': nir_band.flatten(),
    'red_band': red_band.flatten(),
    'green_band': green_band.flatten(),
    'red_edge_band': red_edge_band.flatten(),
    'ndvi': ndvi.flatten(),
    'savi': savi.flatten(),
    'evi2': evi2.flatten(),
    'ndre': ndre.flatten(),
    'gndvi': gndvi.flatten()
}

df = pd.DataFrame(data)

# Guardar el DataFrame en un archivo CSV
df.to_csv('mi_dataframe1.csv', index=False)

# Imprimir las primeras filas del DataFrame
print(df.head())

```

```python
import numpy as np
from matplotlib.image import imread
from sklearn.cluster import KMeans

# Cargar la imagen
image = imread(r"J:\proyecto\codigos\dnir.tif")

# Obtener las dimensiones de la imagen
height, width = image.shape

# Reshape la imagen para tener una matriz de píxeles
pixels = image.reshape(-1, 1)

# Definir el número de clusters (2 en este caso para separar cultivo y suelo)
n_clusters = 2

# Aplicar K-Means para agrupar los píxeles
kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=0)
kmeans.fit(pixels)

# Obtener las etiquetas de los clusters asignados a cada píxel
labels = kmeans.labels_

# Asignar etiquetas binarias (0 para suelo y 255 para cultivo)
segmented_image = np.zeros_like(pixels, dtype=np.uint16)
cultivo_label = np.argmax(np.bincount(labels))  # Determinar la etiqueta del cultivo
segmented_image[labels == cultivo_label] = 65535  # Asignar 255 a los píxeles del cultivo

# Reshape la imagen segmentada a las dimensiones originales
segmented_image = segmented_image.reshape(height, width)

# Mostrar la imagen segmentada
import matplotlib.pyplot as plt
plt.imshow(segmented_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import numpy as np

# Convertir la imagen binaria en una matriz NumPy
matriz_binaria = np.asarray(imagen_binaria, dtype=np.uint16)
```

```python
# Definir los valores correspondientes a los píxeles de cultivo y suelo
valor_cultivo = 65535
valor_suelo = 0

# Crear una matriz de etiquetas inicializada con ceros
etiquetas = np.zeros_like(matriz_binaria, dtype=np.uint16)

# Recorrer la matriz binaria y asignar etiquetas según el valor de cada píxel
etiquetas[matriz_binaria == valor_cultivo] = 1
etiquetas[matriz_binaria == valor_suelo] = 0

# Verificar las dimensiones y el tipo de datos de la matriz de etiquetas
print("Dimensiones de la matriz de etiquetas:", etiquetas.shape)
print("Tipo de datos de la matriz de etiquetas:", etiquetas.dtype)
```
