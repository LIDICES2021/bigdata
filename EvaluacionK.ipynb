# Evaluacion k

```python
import tifffile
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import cv2
from sklearn.metrics import silhouette_score

# Ruta de la imagen TIFF 
ruta_imagen1 = "dnir.tif"

# Carga la imagen TIFF
nir_image = tifffile.imread(ruta_imagen1)
```

```python
# Obtener las dimensiones de la imagen
height, width = nir_image.shape

# Reshape la imagen para tener una matriz de pxeles
pixels = nir_image.reshape(-1, 1)

# Definir el nmero de clusters (2 en este caso para separar cultivo y suelo)
n_clusters = 2
# Aplicar K-Means para agrupar los pxeles
kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=0)
kmeans.fit(pixels)
# Obtener las etiquetas de los clusters asignados a cada pxel
labels = kmeans.labels_

# Asignar etiquetas binarias (0 para cultivo y 65535 para suelo)
segmented_image = np.zeros_like(pixels, dtype=np.uint16)
cultivo_label = np.argmin(np.bincount(labels))  # Determinar la etiqueta del cultivo
segmented_image[labels == cultivo_label] = 65535  # Asignar 65535 a los pxeles del cultivo

# Reshape la imagen segmentada a las dimensiones originales
segmented_image = segmented_image.reshape(height, width)

# Guardar el resultado del NDVI en un archivo TIFF
tifffile.imsave('segmented_K-means_k=2.tif', segmented_image)

# Mostrar la imagen segmentada
plt.imshow(segmented_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Carga la imagen TIFF
image = tifffile.imread("dnir.tif")

# Obtener las dimensiones de la imagen
height, width = image.shape

# Carga las imgenes segmentadas con diferentes valores de k


image_k3 = cv2.imread('segmented_K-means_k=3.tif', cv2.IMREAD_GRAYSCALE)
image_k4 = cv2.imread('segmented_K-means_k=4.tif', cv2.IMREAD_GRAYSCALE)
image_k5 = cv2.imread('segmented_K-means_k=5.tif', cv2.IMREAD_GRAYSCALE)
image_k6 = cv2.imread('segmented_K-means_k=6.tif', cv2.IMREAD_GRAYSCALE)
image_k7 = cv2.imread('segmented_K-means_k=7.tif', cv2.IMREAD_GRAYSCALE)
image_k8 = cv2.imread('segmented_K-means_k=8.tif', cv2.IMREAD_GRAYSCALE)
image_k9 = cv2.imread('segmented_K-means_k=9.tif', cv2.IMREAD_GRAYSCALE)

# Verifica que las imgenes se hayan cargado correctamente
if any(img is None for img in [image_k3, image_k4, image_k5, image_k6, image_k7, image_k8, image_k9]):
    print("Error al cargar una o ms imgenes segmentadas.")
    exit()

# Lista de imgenes segmentadas
segmented_images = [image_k3, image_k4, image_k5, image_k6, image_k7, image_k8, image_k9]

# Lista de valores de k correspondientes a cada imagen segmentada
k_values = [3, 4, 5, 6, 7, 8, 9]

puntos = image.reshape(-1, 1)


# Inicializar una lista para almacenar la WSS en cada iteracin
wss_values = []
centroides = [0,0]

for i in range(len(segmented_images)):
    flattened_image = segmented_images[i].reshape(-1, 1)
    etiquetas = np.where(flattened_image == 255, 1, 0)    

    # Calcular los nuevos centroides como el promedio de los puntos asignados a cada cluster
    for k in range(2):
        centroides[k] = np.mean(puntos[etiquetas == k])

    # Calcular la WSS en esta iteracin
    wss = np.sum(np.min(np.abs(puntos - centroides) ** 2, axis=1))
    wss_values.append(wss)
```

```python
wss_values
```

```python
# Graficar SSE vs k
plt.plot(k_values, wss_values, marker='o')
plt.xlabel("Number of clusters (K)", style="italic")
plt.ylabel('SSW')
plt.show()
```

```python
import cv2
import numpy as np

# Carga la imagen en escala de grises
imagen = cv2.imread('dnir.tif', cv2.IMREAD_UNCHANGED)

# Aplica el mtodo de Otsu para obtener la imagen binaria segmentada
_, imagen_binaria = cv2.threshold(imagen, 0, 65535, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Convierte la imagen binaria en un array de puntos
puntos = np.argwhere(imagen_binaria == 65535)

# Imprime algunos valores para ver si hay algo inusual
print("Imagen binaria:")
print(imagen_binaria)
print("\nPuntos:")
print(puntos)

# Calcula el centroide de los puntos en la imagen binaria
centroide = np.mean(puntos, axis=0)

# Calcula la Within Sum of Squares (WSS)
wss = np.sum((puntos - centroide) ** 2)

print("\nWithin Sum of Squares (WSS):", wss)
```

```python
import cv2
import numpy as np
from sklearn.cluster import KMeans

# Carga la imagen en escala de grises
imagen = cv2.imread('dnir.tif', cv2.IMREAD_UNCHANGED)

# Convierte la imagen en un array de puntos
puntos = np.argwhere(imagen > 0)  # Solo consideramos los puntos distintos de cero

# Ajusta K-Means con k=4
kmeans = KMeans(n_clusters=4, n_init=10, random_state=0)
kmeans.fit(puntos)

# Calcula el WSS para k=4
wss_k4 = kmeans.inertia_

print("Within Sum of Squares (WSS) for k=4:", wss_k4)
```
