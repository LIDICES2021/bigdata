```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar las dos imágenes de 16 bits
imagen1 = cv2.imread('imagen1.tif', cv2.IMREAD_UNCHANGED)
imagen2 = cv2.imread('imagen2.tif', cv2.IMREAD_UNCHANGED)

# Escalar los valores de píxeles al rango de 0 a 255
imagen1_scaled = np.uint8(imagen1 / np.max(imagen1) * 255)
imagen2_scaled = np.uint8(imagen2 / np.max(imagen2) * 255)

# Detectar puntos de interés y descriptores en ambas imágenes
sift = cv2.SIFT_create()
keypoints1, descriptors1 = sift.detectAndCompute(imagen1_scaled, None)
keypoints2, descriptors2 = sift.detectAndCompute(imagen2_scaled, None)

# Establecer el matcher de puntos
matcher = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)

# Encontrar las correspondencias entre los descriptores de ambas imágenes
matches = matcher.match(descriptors1, descriptors2)

# Ordenar las correspondencias por su distancia
matches = sorted(matches, key=lambda x: x.distance)

# Extraer las ubicaciones de los puntos de interés en ambas imágenes
pts1 = np.float32([keypoints1[m.queryIdx].pt for m in matches]).reshape(-1, 1, 2)
pts2 = np.float32([keypoints2[m.trainIdx].pt for m in matches]).reshape(-1, 1, 2)

# Calcular la matriz de transformación afinada
M, _ = cv2.findHomography(pts1, pts2, cv2.RANSAC, 5.0)

# Aplicar la transformación a la imagen 1 para alinearla con la imagen 2
imagen1_aligned = cv2.warpPerspective(imagen1_scaled, M, (imagen2_scaled.shape[1], imagen2_scaled.shape[0]))

# Mostrar las imágenes alineadas
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(imagen1_aligned, cmap='gray')
ax1.set_title('Imagen 1 alineada')
ax2.imshow(imagen2_scaled, cmap='gray')
ax2.set_title('Imagen 2')
plt.show()
```
