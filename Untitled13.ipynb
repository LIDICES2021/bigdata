```python
import matplotlib.pyplot as plt

# Datos de ejemplo (valores de SSE y valores de k)
sse = [250, 150, 100, 80, 70, 65, 60]
k_values = [3, 4, 5, 6, 7, 8, 9]

# Configuración del gráfico
plt.figure(figsize=(8, 6))  # Tamaño del gráfico
plt.title('Gráfico del método del codo')  # Título del gráfico
plt.xlabel('Valor de k')  # Etiqueta del eje x
plt.ylabel('SSE')  # Etiqueta del eje y

# Creación del gráfico de dispersión
plt.plot(k_values, sse, marker='o')

# Anotación del punto de codo
plt.annotate('Codo', xy=(4, 70), xytext=(4, 150),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Mostrar el gráfico
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen segmentada en formato .tif
imagen_segmentada = cv2.imread('segmented_imageOtsu1.tif', 0)

# Calcular el histograma
histograma = cv2.calcHist([imagen_segmentada], [0], None, [256], [0, 256])

# Crear el gráfico del histograma
plt.figure(figsize=(8, 6))  # Tamaño del gráfico
plt.title('Histograma de la imagen segmentada')  # Título del gráfico
plt.xlabel('Intensidad de píxel')  # Etiqueta del eje x
plt.ylabel('Frecuencia')  # Etiqueta del eje y
plt.plot(histograma, color='black')

# Mostrar el gráfico del histograma
plt.show()
```
