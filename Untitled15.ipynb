```python
import cv2
import numpy as np

def calculate_ssw(thresholded_image):
    # Etiquetar los clústeres en la imagen segmentada
    _, labels = cv2.connectedComponents(thresholded_image)

    # Calcular los centroides de los clústeres y el número de clústeres (k)
    k = np.max(labels)
    centroids = [np.mean(np.argwhere(labels == i), axis=0) for i in range(1, k + 1)]

    # Calcular el SSW
    ssw = 0
    for i in range(1, k + 1):
        cluster_points = np.argwhere(labels == i)
        centroid = centroids[i - 1]
        distances = np.linalg.norm(cluster_points - centroid, axis=1)
        ssw += np.sum(distances ** 2)

    return ssw

# Cargar la imagen en formato .tif y convertir a escala de grises
image = cv2.imread('dnir.tif', cv2.IMREAD_GRAYSCALE)

# Aplicar el método de Otsu para obtener el umbral óptimo
_, thresholded_image = cv2.threshold(image, 0, 65535, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Calcular la métrica SSW en la imagen segmentada
ssw_value = calculate_ssw(thresholded_image)
print("Sum of Squared Within (SSW):", ssw_value)
```

```python
import numpy as np
from matplotlib.image import imread
from sklearn.cluster import KMeans
import tifffile

def calculate_ssw(image, labels, centroids):
    k = len(centroids)
    ssw = 0
    for i in range(k):
        cluster_points = image[labels == i]
        centroid = centroids[i]
        distances = np.linalg.norm(cluster_points - centroid)
        ssw += np.sum(distances ** 2)
    return ssw

# Cargar la imagen
image = imread(r"dnir.tif")

# Obtener las dimensiones de la imagen
height, width = image.shape

# Reshape la imagen para tener una matriz de píxeles
pixels = image.reshape(-1, 1)

# Definir el número de clusters (4 en este caso)
n_clusters = 4

# Aplicar K-Means para agrupar los píxeles
kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=0)
kmeans.fit(pixels)

# Obtener las etiquetas de los clusters asignados a cada píxel
labels = kmeans.labels_

# Calcular el SSW para la segmentación con K-means
ssw_kmeans = calculate_ssw(pixels, labels, kmeans.cluster_centers_)

# Asignar etiquetas binarias (0 para cultivo y 65535 para suelo)
segmented_image = np.zeros_like(pixels, dtype=np.uint16)
cultivo_label = np.argmin(np.bincount(labels))  # Determinar la etiqueta del cultivo
segmented_image[labels == cultivo_label] = 65535  # Asignar 65535 a los píxeles del cultivo

# Reshape la imagen segmentada a las dimensiones originales
segmented_image = segmented_image.reshape(height, width)

# Guardar el resultado del NDVI en un archivo TIFF
tifffile.imsave('segmented_image_result4.tif', segmented_image)

# Mostrar la imagen segmentada
import matplotlib.pyplot as plt
plt.imshow(segmented_image, cmap='gray')
plt.axis('off')
plt.show()

print("Sum of Squared Within (SSW) - K-means:", ssw_kmeans)
```
