```python
import os
from tifffile import imread

# Ruta de la carpeta que contiene las imágenes binarias
ruta_carpeta_imagenes_binarias = "img_KITTLER_NIR"

# Obtener la lista de nombres de archivo de las imágenes binarias
nombres_imagenes_binarias = os.listdir(ruta_carpeta_imagenes_binarias)

# Recorrer todas las imágenes binarias y mostrar sus valores
for nombre_imagen_binaria in nombres_imagenes_binarias:
    ruta_imagen_binaria = os.path.join(ruta_carpeta_imagenes_binarias, nombre_imagen_binaria)
    imagen_binaria = imread(ruta_imagen_binaria)
    
    print(f"Valores de la imagen binaria '{nombre_imagen_binaria}':")
    print(imagen_binaria)
    print()
```
