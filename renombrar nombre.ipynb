```python
import os

# Ruta del directorio donde se encuentran las imÃ¡genes
directorio = "img_KITTLER_NIR"

# Obtener la lista de archivos en el directorio
archivos = os.listdir(directorio)

# Recorrer cada archivo en el directorio
for archivo in archivos:
    # Verificar si el archivo contiene la palabra "_otsu"
    if "_kittler" in archivo:
        # Generar el nuevo nombre del archivo sin la palabra "_otsu"
        nuevo_nombre = archivo.replace("_kittler", "")
        
        # Ruta completa del archivo antiguo
        ruta_antigua = os.path.join(directorio, archivo)
        
        # Ruta completa del archivo nuevo
        ruta_nueva = os.path.join(directorio, nuevo_nombre)
        
        # Renombrar el archivo
        os.rename(ruta_antigua, ruta_nueva)
        print(f"Archivo renombrado: {archivo} -> {nuevo_nombre}")
```

```python
from PIL import Image
import os

# Ruta del directorio donde se encuentran las imÃ¡genes
directorio = r"img/REG"

# TamaÃ±o de redimensionamiento deseado
nuevo_ancho = 1280
nuevo_alto = 960

# Obtener la lista de archivos en el directorio
archivos = os.listdir(directorio)

# Recorrer cada archivo en el directorio
for archivo in archivos:
    # Ruta completa de la imagen
    ruta_imagen = os.path.join(directorio, archivo)
    
    try:
        # Abrir la imagen TIFF multiespectral
        imagen = Image.open(ruta_imagen)
        
        # Redimensionar la imagen
        imagen = imagen.resize((nuevo_ancho, nuevo_alto))
        
        # Guardar la imagen redimensionada
        imagen.save(ruta_imagen)
        
        print(f"Imagen redimensionada: {archivo}")
    except Exception as e:
        print(f"Error al redimensionar la imagen {archivo}: {str(e)}")
```

```python
import cv2
import os

# Ruta del directorio donde se encuentran las imÃ¡genes
directorio = r"img/NIR"

# Obtener la lista de archivos en el directorio
archivos = os.listdir(directorio)

# Recorrer cada archivo en el directorio
for archivo in archivos:
    # Ruta completa de la imagen
    ruta_imagen = os.path.join(directorio, archivo)
    
    try:
        # Cargar la imagen TIFF multiespectral en escala de grises
        imagen = cv2.imread(ruta_imagen, cv2.IMREAD_GRAYSCALE)
        
        # Normalizar los valores de pÃ­xeles entre 0 y 255
        imagen_normalizada = cv2.normalize(imagen, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
        
        # Guardar la imagen normalizada
        cv2.imwrite(ruta_imagen, imagen_normalizada)
        
        print(f"Imagen normalizada: {archivo}")
    except Exception as e:
        print(f"Error al normalizar la imagen {archivo}: {str(e)}")
```
