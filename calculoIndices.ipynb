```python
import rasterio

# Rutas de las imágenes multiespectrales
nir_image_path = "CAL_NIR.tif"
red_image_path = "CAL_RED.tif"
green_image_path = "CAL_GRE.tif"
red_edge_image_path = "CAL_REG.tif"

# Abrir las imágenes
with rasterio.open(nir_image_path) as nir_src, \
        rasterio.open(red_image_path) as red_src, \
        rasterio.open(green_image_path) as green_src, \
        rasterio.open(red_edge_image_path) as red_edge_src:
    # Leer las bandas espectrales
    nir_band = nir_src.read(1)
    red_band = red_src.read(1)
    green_band = green_src.read(1)
    red_edge_band = red_edge_src.read(1)

```

```python
import numpy as np

# Convertir los datos de la imagen a float32 en el rango 0-1
nir_band = nir_band.astype(np.float32) / 65535.0
red_band = red_band.astype(np.float32) / 65535.0
green_band = green_band.astype(np.float32) / 65535.0
red_edge_band = red_edge_band.astype(np.float32) / 65535.0
```

```python
# Calcular el NDVI
ndvi = (nir_band - red_band) / (nir_band + red_band)

from skimage import io, transform
# Guardar el resultado en una imagen
io.imsave(r'ndvi_image.tif', ndvi)

print("Mapa NDVI:")
print(ndvi)
```

```python
import matplotlib.pyplot as plt

# Visualizar el mapa NDVI
plt.imshow(ndvi, cmap='RdYlGn')
plt.colorbar(label='NDVI')
plt.title('Mapa NDVI')
plt.show()
```
