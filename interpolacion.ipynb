```python
import rasterio
from rasterio.windows import Window
import numpy as np

# Rutas de las imágenes TIFF
ruta_imagen1 = 'NIR.tif'
ruta_imagen2 = 'RED.tif'
ruta_imagen_salida = 'ruta_de_la_imagen_salida.tif'

# Cargar las imágenes TIFF
with rasterio.open(ruta_imagen1) as src1:
    imagen1 = src1.read(1)

with rasterio.open(ruta_imagen2) as src2:
    imagen2 = src2.read(1)

# Dimensiones de la imagen de salida (tomando las dimensiones máximas)
alto_deseado = max(imagen1.shape[0], imagen2.shape[0])
ancho_deseado = max(imagen1.shape[1], imagen2.shape[1])

# Crear la imagen de salida con las dimensiones adecuadas
imagen_salida = np.zeros((alto_deseado, ancho_deseado), dtype=np.uint16)

# Copiar imagen1 en la posición deseada de la imagen de salida
imagen_salida[:imagen1.shape[0], :imagen1.shape[1]] += imagen1

# Calcular la diferencia de dimensiones entre imagen2 y la imagen de salida
dif_alto = imagen_salida.shape[0] - imagen2.shape[0]
dif_ancho = imagen_salida.shape[1] - imagen2.shape[1]

# Crear una ventana de recorte para imagen2
window = Window(dif_ancho, dif_alto, imagen2.shape[1], imagen2.shape[0])

# Leer la parte relevante de imagen2 y copiarla en la posición deseada de la imagen de salida
imagen_salida[window.row_off:window.row_off + window.height, window.col_off:window.col_off + window.width] += imagen2

# Guardar la imagen de salida como TIFF
with rasterio.open(ruta_imagen_salida, 'w', driver='GTiff', width=ancho_deseado, height=alto_deseado, count=1, dtype='uint16') as dst:
    dst.write(imagen_salida, 1)

```

```python
import rasterio
import numpy as np

# Rutas de las imágenes TIFF
ruta_imagen1 = 'NIR.tif'
ruta_imagen2 = 'RED.tif'
ruta_imagen_salida = 'ruta_de_la_imagen_salida3.tif'

# Cargar las imágenes TIFF
with rasterio.open(ruta_imagen1) as src1:
    imagen1 = src1.read(1)

with rasterio.open(ruta_imagen2) as src2:
    imagen2 = src2.read(1)

# Dimensiones de la imagen de salida (tomando la dimensión máxima)
alto_deseado = max(imagen1.shape[0], imagen2.shape[0])
ancho_deseado = max(imagen1.shape[1], imagen2.shape[1])

# Redimensionar ambas imágenes a las dimensiones de la imagen de salida
imagen1_resized = np.zeros((alto_deseado, ancho_deseado), dtype=np.uint16)
imagen2_resized = np.zeros((alto_deseado, ancho_deseado), dtype=np.uint16)
imagen1_resized[:imagen1.shape[0], :imagen1.shape[1]] = imagen1
imagen2_resized[:imagen2.shape[0], :imagen2.shape[1]] = imagen2

# Superponer las imágenes utilizando una suma ponderada
peso_imagen1 = 0.5  # Ajustar el peso según tus preferencias
peso_imagen2 = 0.5

imagen_salida = peso_imagen1 * imagen1_resized + peso_imagen2 * imagen2_resized

# Guardar la imagen de salida como TIFF
with rasterio.open(ruta_imagen_salida, 'w', driver='GTiff', width=ancho_deseado, height=alto_deseado, count=1, dtype='uint16') as dst:
    dst.write(imagen_salida, 1)
```
