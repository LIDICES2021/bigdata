```python
import cv2
import numpy as np
import os

# Directorio de las imágenes
directorio = r"img\NIR"

# Obtener lista de archivos en el directorio
archivos = os.listdir(directorio)

# Recorrer cada archivo de imagen
for archivo in archivos:
    if archivo.endswith(".TIF"):
        # Leer la imagen
        ruta_imagen = os.path.join(directorio, archivo)
        imagen = cv2.imread(ruta_imagen, cv2.IMREAD_UNCHANGED)
        
        # Normalizar los valores de píxeles entre 0 y 255
        imagen_normalizada = cv2.normalize(imagen, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
        
        # Aplicar el método de Otsu
        _, imagen_binaria = cv2.threshold(imagen_normalizada, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        
        # Guardar la imagen binaria resultante
        nombre_salida = archivo.replace(".TIF", "_otsu.TIF")
        ruta_salida = os.path.join(directorio, nombre_salida)
        cv2.imwrite(ruta_salida, imagen_binaria)
        
        print(f"Procesada imagen: {archivo}")

print("Proceso completado.")
```

```python
import cv2
import numpy as np
import os

# Directorio de las imágenes
directorio = r"img\NIR"

# Obtener lista de archivos en el directorio
archivos = os.listdir(directorio)

# Recorrer cada archivo de imagen
for archivo in archivos:
    if archivo.endswith(".TIF"):
        # Leer la imagen
        ruta_imagen = os.path.join(directorio, archivo)
        imagen = cv2.imread(ruta_imagen, cv2.IMREAD_UNCHANGED)
        
        # Convertir la imagen a tipo uint16
        imagen_uint16 = imagen.astype(np.uint16)
        
        # Aplicar el método de Kittler e Illingworth
        _, imagen_binaria = cv2.threshold(imagen_uint16, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        
        # Convertir la imagen binaria a tipo uint8
        imagen_binaria = imagen_binaria.astype(np.uint8)
        
        # Guardar la imagen binaria resultante
        nombre_salida = archivo.replace(".TIF", "_kittler.TIF")
        ruta_salida = os.path.join(directorio, nombre_salida)
        cv2.imwrite(ruta_salida, imagen_binaria)
        
        print(f"Procesada imagen: {archivo}")

print("Proceso completado.")
```
