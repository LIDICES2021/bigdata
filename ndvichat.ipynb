```python
import numpy as np
from skimage import io, transform

# Cargar las imágenes multiespectrales (rojo y NIR)
red_image = io.imread(r'RED.tif')
nir_image = io.imread(r'NIR.tif')

# Convertir las imágenes en arrays de tipo float
red_image = red_image.astype(float)
nir_image = nir_image.astype(float)

# Calcular el NDVI
ndvi = (nir_image - red_image) / (nir_image + red_image)

# Guardar el resultado en una imagen
io.imsave(r'ndvi_image.tif', ndvi)

# Visualizar el resultado
io.imshow(ndvi)
io.show()
```
